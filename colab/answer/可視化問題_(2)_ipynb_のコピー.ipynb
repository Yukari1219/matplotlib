## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

```python
data = pd.read_csv('https://raw.githubusercontent.com/Yukari1219/matplotlib/master/colab/quizzes/matplotlib_quiz/hist_data.csv')
a_data = data['aデータ']
b_data = data['bデータ']
plt.hist(a_data, bins = 30, alpha = 0.5, color = 'orange', label = 'aデータ')
plt.hist(b_data, bins = 30, alpha = 0.5, color = 'green', label = 'bデータ')

plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

```python
data = pd.read_csv('https://raw.githubusercontent.com/Yukari1219/matplotlib/master/colab/quizzes/matplotlib_quiz/scatter_data.csv')
class0 = data[data['class'] == 0]
class1 = data[data['class'] == 1]
class2 = data[data['class'] == 2]

class0_x = class0['length (cm)']
class0_y = class0['width (cm)']
class1_x = class1['length (cm)']
class1_y = class1['width (cm)']
class2_x = class2['length (cm)']
class2_y = class2['width (cm)']

plt.scatter(class0_x, class0_y, color = 'red', label = 'class0')
plt.scatter(class1_x, class1_y, color = 'blue', label = 'class1')
plt.scatter(class2_x, class2_y, color = 'green', label = 'class2')

plt.title('クラス０～２のlengthとwidthに関する散布図')
plt.xlabel('length(cm)')
plt.ylabel('width(cm)')
plt.grid()
plt.legend()
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Yukari1219/matplotlib/master/colab/quizzes/matplotlib_quiz/line_data.csv')
data_period = data['期間']
data_num = data['件数']


plt.plot(data_period, data_num, marker = 'o')
plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.xticks(rotation = 30)
plt.ylabel('件数')
plt.grid()
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Yukari1219/matplotlib/master/colab/quizzes/matplotlib_quiz/line_data.csv')
data_cost = data['費用']
plt.bar(data_period, data_cost)
plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.xticks(rotation = 30)
plt.ylabel('費用(円)')
plt.grid()
```
